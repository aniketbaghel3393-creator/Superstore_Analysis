Make a Discount like (No Discount, Low Discount, Medium Discount, High Discount) from using if Condition.

```python
superstore["Discount"]
```

```python
def discount_category(Discount):
    if Discount == 0:
      print("No Discount")

    elif Discount >= 0.2:
      print("Low Discount")

    elif Discount >= 0.5:
      print("Medium Discount")
    
    else:
      print("High Discount")
```

```python
superstore["Discount"].apply(discount_category)
```
